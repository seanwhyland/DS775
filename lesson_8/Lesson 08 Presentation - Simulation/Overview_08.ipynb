## Learning Outcomes

The student will be able to:

* define of the state of a system (e.g., the number of customers in a queueing system).
* identify the possible states of the system that can occur.
* identify the possible events (e.g., arrivals and service completions in a queueing system) that would change the state of the system.
* create a simulation clock, located at some address in the simulation program, that will record the passage of (simulated) time.
* generate random outcomes for continuous and discrete events.
* identify and track state transitions that are generated by the various kinds of events.
* use a computer to simulate a physical process or series of events that involve random inputs and outcomes and capture output for variables of interest.
* summarize the results of simulation variables with descriptive statistics, probabilities, and confidence intervals as needed.
* perform parameter analysis on a simulation.
* create a trend chart from a parameter analysis for a simulation.
* optimize a linear program within a simulation.
